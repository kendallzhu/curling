```python
from dataclasses import dataclass
import numpy as np
```

```python
class Gate:
    pass

class Inputs(Gate):
    pass

class Linear(Gate):
    def __init__(self, weights: np.array):
        self.weights = weights
        n_out, n_in = self.weights.shape
        self.inputs = np.zeros(n_in)
        self.output_grad = np.zeros(n_out)

    def forward(self, inputs: np.array):
        self.inputs = inputs
        return self.weights @ inputs

    def backward(self, output_grad: np.array):
        self.output_grad = output_grad
        return self.weights.T @ output_grad

    def update(self, learning_rate: float, regularization: float):
        n_out, n_in = self.weights.shape
        self.weights -= (
            learning_rate * self.output_grad.reshape((n_out, 1)) @ self.inputs.reshape((1,n_in))
            + regularization * self.weights
        )        
        
class Sum(Gate):
    def forward(self, inputs: np.array): # inputs: m x n
        self.inputs = inputs
        return np.sum(self.inputs)
    def backward(self, output_grad: float):
        return np.ones(self.inputs.shape) * output_grad

class Max0(Gate):
    def forward(self, inputs: np.array):
        self.inputs = inputs
        return np.fmax(inputs, 0)
    def backward(self, output_grad: np.array):
        self.output_grad = output_grad
        return np.where(self.inputs > 0, 1, 0) * self.output_grad

class MapTo01(Gate):
    def forward(self, inputs: np.array):
        self.inputs = inputs
        return np.exp(inputs) / (1 + np.exp(inputs))
    def backward(self, output_grad: np.array):
        self.output_grad = output_grad
        return (np.exp(self.inputs) / (1 + np.exp(self.inputs)) ** 2) * self.output_grad
        
class Vertex:
    pass
    
@dataclass
class MultiInputVertex(Vertex):
    input_vertices: list[Vertex]
    gate: Gate

    
@dataclass 
class SingleInputVertex(Vertex):
    input_vertex: Vertex
    gate: Gate
    def __init__(self, input_vertex, gate):
        self.input_vertex = input_vertex
        self.gate = gate
    def forward(self, inputs: np.array):
        return self.gate.forward(self.input_vertex.forward(inputs))

@dataclass
class StartVertex(Vertex):
    def __init__(self):
        pass
    def forward(self, inputs: np.array):
        return inputs

def normalize(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

class SquaredErrorLoss:
    def get_loss(self, x, y):
        return (x - y) ** 2
    def backward(self, x, y):
        return 2 * (x - y)
```

```python
n = 10_000
theta = 6 * np.pi * np.random.random(n).reshape((n, 1))
r = 6 * np.pi * np.random.random(n).reshape((n, 1))
y = np.abs((r - theta) % (2 * np.pi)).flatten() < np.pi
X = r * np.concatenate([np.cos(theta), np.sin(theta)], axis=1)
```

```python
l1 = Linear(1 / (X.std() * np.sqrt(2)) * np.random.normal(size=(10, 2)))
act1 = Max0()
l2 = Linear(1 / np.sqrt(10) * np.random.normal(size=(1, 10)))
act2 = MapTo01()   

gates = [l1, act1, l2, act2]
vertices = [StartVertex()]
for g in gates:
    vertices.append(SingleInputVertex(input_vertex=vertices[-1], gate=g))
nn = vertices[-1]
```
